```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
embeddings_dict = {}
with open("glove.6B.300d.txt", 'r',  encoding="utf8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
```

```python
df_single = pd.read_csv("train/lcp_single_train.tsv", sep='\t')
df_multi = pd.read_csv("train/lcp_multi_train.tsv", sep='\t')
```

```python
df_single_bible = df_single[df_single['corpus']=='bible']
df_single_europarl = df_single[df_single['corpus']=='europarl']
df_single_biomed = df_single[df_single['corpus']=='biomed']
```

```python
tokens = df_single_bible['token'].values
complexity = df_single_bible['complexity'].values
```

```python
x_train = []
glove_complexity = []
glove_tokens = []
for i in range(len(tokens)):
    if tokens[i].lower() in embeddings_dict.keys():
        glove_complexity.append(complexity[i])
        glove_tokens.append(tokens[i].lower())
        x_train.append(embeddings_dict[tokens[i].lower()])

x_train = np.array(x_train, dtype=np.float32)
y_train = np.array(glove_complexity, dtype=np.float32)
y_train = y_train.reshape(-1, 1)
```

```python
import torch
from torch.autograd import Variable
class linearRegression(torch.nn.Module):
    def __init__(self, inputSize, outputSize):
        super(linearRegression, self).__init__()
        self.linear = torch.nn.Linear(inputSize, outputSize)

    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
inputDim = x_train.shape[1]        # takes variable 'x' 
outputDim = 1       # takes variable 'y'
learningRate = 5e-3 
epochs = 1000

model = linearRegression(inputDim, outputDim)

criterion = torch.nn.MSELoss() 
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)
```

```python
for epoch in range(epochs):
    inputs = Variable(torch.from_numpy(x_train))
    labels = Variable(torch.from_numpy(y_train))

    # Clear gradient buffers because we don't want any gradient from previous epoch to carry forward, dont want to cummulate gradients
    optimizer.zero_grad()

    # get output from the model, given the inputs
    outputs = model(inputs)

    # get loss for the predicted output
    loss = criterion(outputs, labels)
#     print(loss)
    # get gradients w.r.t to parameters
    loss.backward()

    # update parameters
    optimizer.step()

    print('epoch {}, loss {}'.format(epoch, loss.item()))
```

```python
with torch.no_grad(): # we don't need gradients in the testing phase
    predicted = model(Variable(torch.from_numpy(x_train))).data.numpy()
```

```python
def mae(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs(actual - pred))
print(mae(y_train,predicted))
```
